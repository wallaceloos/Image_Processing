```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def skeletons(img):
    k = 0
    x0 = np.copy(img)    
    b = np.ones((3,3), dtype="uint8")

    openingA = cv2.morphologyEx(x0, cv2.MORPH_OPEN, b)
    a = abs(x0 - openingA)
    
    while np.max(x0) > 0:
        erosionA = cv2.erode(x0,b)
        openingA = cv2.morphologyEx(erosionA, cv2.MORPH_OPEN, b)
        x0 = erosionA
        a = cv2.bitwise_or(a, abs(erosionA - openingA))
        #a = a.astype("uint8")
        k = k+1
    print ("K:",k)
    return a,k
```

```python
img = cv2.imread("convex_hull1.png",0)
sk,k = skeletons(img)

plt.figure(figsize=(20,20))

plt.subplot(1,2,1)
plt.title('Original Image',fontsize=22)
plt.axis("off")
plt.imshow(img,'gray')

plt.subplot(1,2,2)
plt.title('Skeleton + Dilation',fontsize=22)
plt.axis("off")
plt.imshow(cv2.dilate(sk,np.ones((3,3))),'gray')


plt.show()
```
